```python
import datadigest
from datadigest.utils import pd2spark
from datadigest.base import *
```

```python
ints = [True, False, 
        np.iinfo('int8').min-1, np.iinfo('int8').max+1,
        np.iinfo('int16').min-1, np.iinfo('int16').max+1,
        np.iinfo('int32').min-1, np.iinfo('int32').max+1,
        np.iinfo('int64').min, np.iinfo('int64').max,
        0
       ]
df = pd.DataFrame({
    # floats and ints can change automatically based on platform support
    
    'a_int8': pd.Series(ints).astype(np.int8),
    'a_int16': pd.Series(ints).astype(np.int16),
    'a_int32': pd.Series(ints).astype(np.int32),
    'a_int64': pd.Series(ints).astype(np.int64),
    'a_int': pd.Series(ints).astype(np.int),
    'a_int0': pd.Series(ints).astype(np.int0),
    'a_int_': pd.Series(ints).astype(np.int_),
})

df
df.dtypes
df = pd2spark(df)
```

```python
df.show()
```

```python
a = describe_1d(df, 'a_int8')
b = describe_1d(df, 'a_int16')
```
